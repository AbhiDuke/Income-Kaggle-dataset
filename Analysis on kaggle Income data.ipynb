```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_x,scaled_y, test_size=0.70,stratify=scaled_y )
from sklearn.ensemble import GradientBoostingClassifier
gradt=GradientBoostingClassifier(random_state=0,max_depth=1)
gradt.fit(X_train,y_train)
pred_grad = gradt.predict(X_test)
print ('Accuracy on training set:{:.3f}'.format(gradt.score(X_train,y_train)))
print ('Accuracy on testing set:{:.3f}'.format(gradt.score(X_test,y_test)))
print (confusion_matrix(y_test,pred_grad))
print('\n')
print (classification_report(y_test,pred_grad))
print('\n')
print (accuracy_score(y_test,pred_grad))
```

```python
from sklearn.svm import SVC
svc_m = SVC()
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.70,stratify=y)
svc_m.fit(X_train,y_train)
pred_svm=svc_m.predict(X_test)
from sklearn.grid_search import GridSearchCV
par_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.001,0.0001,0.00001]}
gridserc = GridSearchCV(SVC(),param_grid=par_grid)
gridserc.fit(X_train,y_train)
print ('Accuracy on training set:{:.3f}'.format(gridserc.score(X_train,y_train)))
print ('Accuracy on testing set:{:.3f}'.format(gridserc.score(X_test,y_test)))
print (confusion_matrix(y_test,pred_svm))
print('\n')
print (classification_report(y_test,pred_svm))
print('\n')
print (accuracy_score(y_test,pred_svm))
```

```python
from sklearn.grid_search import GridSearchCV
par_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.001,0.0001,0.00001]}
gridserc = GridSearchCV(SVC(),param_grid=par_grid,verbose=5)
gridserc.fit(X_train,y_train)
gridserc_pred = gridserc.predict(X_test)
print ('Accuracy on training set:{:.3f}'.format(gridserc.score(X_train,y_train)))
print ('Accuracy on testing set:{:.3f}'.format(gridserc.score(X_test,y_test)))
print (confusion_matrix(y_test,gridserc_pred))
print('\n')
print (classification_report(y_test,gridserc_pred))
print('\n')
print (accuracy_score(y_test,gridserc_pred))
```
